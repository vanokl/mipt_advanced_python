```python
import pandas as pd
```

```python
data = pd.read_csv("movies_stats.csv")
year_list = set(range(1950, 2010 + 1))
```

```python
data.head()
```

```python
def production_year(movie_name):
    default_prod_year = 1900

    for year in year_list:
        if str(year) in movie_name:
            return year
        
    return default_prod_year
```

```python
data["year"] = data["title"].apply(lambda x: production_year(x))
```

```python
data.head()
```

```python
data.groupby("year").agg({"rating": "mean"}).sort_values("rating", ascending=False)
```
