
Какой тип(ы) переменные участвует в непрерывный?
- непрерывный

Какая цель у данной переменной в визуализации? 
- показать распределение времени поездки

Почему выбран именно такой тип графика? Аргументируйте
- hist plot удобен для отображения распределния непрервыного типа данных 

```python
df["dropoff_datetime"] = pd.to_datetime(df['dropoff_datetime']).dt.month
df_day_aggr = df.groupby('dropoff_datetime')["id"].count().to_frame().reset_index()

plt.plot(df_day_aggr["dropoff_datetime"], df_day_aggr['id'])
plt.title('Динамика использования такси по месяцам')
plt.xlabel('дата')
plt.ylabel('Количество поездок на такси')
```
